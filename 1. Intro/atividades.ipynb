- Três lados formam um triângulo quando a soma de quaisquer dois lados for maior que o terceiro;
- Triângulo Equilátero: três lados iguais;
- Triângulo Isósceles: quaisquer dois lados iguais;
- Triângulo Escaleno: três lados diferentes;

O preço do litro de diesel é R\$ 2,00 e o preço do litro de etanol é R\$ 1,70. Escreva um programa que leia a quantidade de litros vendidos e o tipo de combustível (E para etanol e D para diesel) e calcule o valor a ser pago pelo cliente. Tenha em mente algumas dicas:

- O do valor do desconto será a multiplicação entre preço do litro, quantidade de litros e o valor do desconto.

````python
litros = float(input('Digite a quantidade de litros: '))
tipo_combustivel = input('Digite o tipo de combustível (E para etanol e D para diesel): ').upper()

if tipo_combustivel == 'E':
    preco_litro = 1.70
    if litros <= 15:
        desconto = 0.02
    else:
        desconto = 0.04
elif tipo_combustivel == 'D':
    preco_litro = 2.00
    if litros <= 15:
        desconto = 0.03
    else:
        desconto = 0.05
  # Caso ocorra um erro na especificação de tipo de combustível,
# consideramos entradas inválidas, e os preços são taxados em 0
else:
    print('Entradas inválidas!')
    preco_litro = 0
    desconto = 0

valor_desconto = preco_litro * litros * desconto
valor_pago = preco_litro * litros - valor_desconto

print(f'O valor a ser pago é de R$ {valor_pago:}')
````

#### Questão 2
Escreva um programa para calcular quantos dias levará para a colônia de uma bactéria A ultrapassar ou igualar a colônia de uma bactéria B, com base nas taxas de crescimento de 3% e 1,5% respectivamente. Considere que a colônia A inicia com 4 elementos e a colônia B com 10 elementos.

````python
bacteria_a = 4
bacteria_b = 10
taxa_a = 0.03
taxa_b = 0.015

if bacteria_a < bacteria_b:
    dias = 0
    while bacteria_a <= bacteria_b:
        bacteria_a += bacteria_a * taxa_a
        bacteria_b += bacteria_b * taxa_b
        dias += 1

print(f'Serão necessários {dias} dias para que a colônia A ultrapasse a colônia B.')
````

#### Questão 3
Para tratar uma quantidade de 15 dados de avaliações de pessoas usuárias de um serviço da empresa, precisamos verificar se as notas são válidas. Caso seja inserido uma nota acima de 5 ou abaixo de 0, repita até que a pessoa usuária insira um valor válido.

````python
notas = []

for i in range(15):
    nota = float(input('Digite uma nota de 0 a 5: '))
    while nota < 0 or nota > 5:
        print('Nota inválida!')
        nota = float(input('Digite uma nota de 0 a 5: '))
    notas.append(nota)

print("\nNotas válidas registradas:")
print(notas)
````

````python
temperaturas = []

print("Digite as temperaturas em Celsius. Digite -273 para encerrar.\n")

for i in range(100):
    temperatura = float(input('Digite uma temperatura em Celsius: '))
    if temperatura == -273:
        break
    temperaturas.append(temperatura)

media = sum(temperaturas) / len(temperaturas)
print(f"\nMédia das temperaturas: {media:.2f}°C")
````

#### Questão 9
Em uma eleição para gerência em uma empresa com 20 funcionários, existem quatro candidatos. Escreva um programa que calcule o vencedor da eleição. A votação ocorreu da seguinte maneira:
- Cada funcionário votou em um dos quatro candidatos (representados pelos números 1, 2, 3 e 4).
- Também foram contabilizados os votos nulos (representado pelo número 5) e os votos em branco (representado pelo número 6).

Ao final da votação, o programa deve exibir o total de votos para cada candidato, o número de votos nulos e o número de votos em branco. Além disso, deve calcular e exibir a porcentagem de votos nulos em relação ao total de votos e a porcentagem de votos em branco em relação ao total de votos.

````python
total_funcionarios = 20

# Contadores de votos
votos = {
    1: 0,  # candidato 1
    2: 0,  # candidato 2
    3: 0,  # candidato 3
    4: 0,  # candidato 4
    5: 0,  # votos nulos
    6: 0   # votos em branco
}

print("Digite o voto de cada funcionário (1 a 6):")
print("1-4: candidatos, 5: voto nulo, 6: voto em branco")

for i in range(1, total_funcionarios + 1):
    while True:
        voto = int(input(f"Voto do funcionário #{i}: "))
        if voto in [1, 2, 3, 4, 5, 6]:
            votos[voto] += 1
            break
        else:
            print("Voto inválido! Digite um número entre 1 e 6.")

total_votos = sum(votos.values())

# Exibe resultados
print("\nResultado da eleição:")
for candidato in range(1, 5):
    print(f"Candidato {candidato}: {votos[candidato]} votos")

print(f"Votos nulos: {votos[5]} votos")
print(f"Votos em branco: {votos[6]} votos")

# Calcula porcentagens
porc_nulos = (votos[5] / total_votos) * 100
porc_brancos = (votos[6] / total_votos) * 100

print(f"\nPorcentagem de votos nulos: {porc_nulos:.2f}%")
print(f"Porcentagem de votos em branco: {porc_brancos:.2f}%")

# Calcula vencedor (entre os candidatos 1 a 4)
vencedor = max(range(1,5), key=lambda c: votos[c])
print(f"\nO vencedor da eleição é o candidato {vencedor} com {votos[vencedor]} votos.")
````

Dica: para calcular o percentual de crescimento usamos a seguinte equação: `100 * (amostra_atual - amostra_passada) / (amostra_passada)`.

````python
bacterias_colonia = [1.2, 2.1, 3.3, 5.0, 7.8, 11.3, 16.6, 25.1, 37.8, 56.9]

porcentagem_crescimento = []

for i in range(1, len(bacterias_colonia)):
    porcentagem = 100 * (bacterias_colonia[i] - bacterias_colonia[i-1]) / bacterias_colonia[i-1]
    porcentagem_crescimento.append(round(porcentagem, 2))

print('Porcentagens de crescimento:')
print(porcentagem_crescimento)
````
